## Introducción

El uso de modelos de aprendizaje automático que clasifiquen texto se ha convertido en un tema de gran interés y está siendo ampliamente utilizado por muchas empresas que, a partir de clasificadores automáticos, reúnen información sobre textos o mensajes escritos en diferentes plataformas como Twitter o Facebook. Al entrenar estos modelo se puede llegar a saber cosas como si un mensaje es racista, si el texto habla de política o tecnología, o si un comentario es positivo o negativo, incluso hasta el idioma en que se encuentra un texto.

Film Junky Union, una nueva comunidad vanguardista para los aficionados de las películas clásicas, está desarrollando un sistema para filtrar y categorizar reseñas de películas. Para lo cual, busca entrenar un modelo que permita detectar las críticas negativas de forma automática y distinguirlas de las críticas positivas. Se cuenta con un conjunto de datos de reseñas de películas de IMDB con leyendas de polaridad, a partir del cual se construirá un modelo para clasificar las reseñas positivas y negativas, utilizando el puntaje F1 para evaluar el rendimiento de los modelos.

## EDA

Ahora procederemos con el análisis exploratorio de los datos. Empezaremos analizando como cambia el número de películas y reseñas a lo largo de los años. Trazaremos gráficos de barras que nos permitan observar la relación existente entre el número de películas y reseñas a lo largo de los años. A su vez, trazaremos un gráfico de líneas para determinar como cambia el número de reseñas por película a través del tiempo.

```python
# Establecemos el número de filas y columnas, así como el tamaño de nuestros subplots.
fig, axs = plt.subplots(2, 1, figsize=(16, 8))

# Eje 0 - Gráfico para la primera fila
ax = axs[0]

# Obtenemos el recuento de valores únicos para cada año y trazamos un gráfico de barras.
dft1 = df_reviews[['tconst', 'start_year']].drop_duplicates() \
    ['start_year'].value_counts().sort_index()
dft1 = dft1.reindex(index=np.arange(dft1.index.min(), max(dft1.index.max(), 2021))).fillna(0)
dft1.plot(kind='bar', ax=ax)
ax.set_title('Número de películas a lo largo de los años')

# Eje 1 - Gráfico para la segunda fila
ax = axs[1]

# Agrupamos por el año y el tipo de reseña, luego obtenemos el número de reseñas positivas y negativas en cada año.
dft2 = df_reviews.groupby(['start_year', 'pos'])['pos'].count().unstack()
dft2 = dft2.reindex(index=np.arange(dft2.index.min(), max(dft2.index.max(), 2021))).fillna(0)

# Trazamos un gráfico de barras
dft2.plot(kind='bar', stacked=True, ax=ax)

# Obtenemos el número de reseñas por año
dft2 = df_reviews['start_year'].value_counts().sort_index()
dft2 = dft2.reindex(index=np.arange(dft2.index.min(), max(dft2.index.max(), 2021))).fillna(0)

# Obtenemos el número de reseñas que una película recibiría cada año
# Trazamos un gráfico de líneas para el número de reseñas por película cada cinco años
dft3 = (dft2/dft1).fillna(0)
axt = ax.twinx()
dft3.reset_index(drop=True).rolling(5).mean().plot(color='orange', label='Reseñas por película (promedio 5 años)', ax=axt)

# Establecemos una leyenda que indique el contenidoe del segundo gráfico
lines, labels = ax.get_legend_handles_labels()
labels = ['0 - Reseñas negativas', '1 - Reseñas positivas']
lines_2, labels_2 = axt.get_legend_handles_labels()
ax.legend(lines + lines_2, labels + labels_2, loc='upper left')
ax.set_title('Número de reseñas a lo largo de los años')

fig.tight_layout()
```

Ahora vamos a ver la distribución del número de reseñas por película, utilizaremos gráficos de barras para visualizar el conteo exacto y KDE para saber cómo puede diferir del conteo exacto.


```python
# Establecemos el número de filas y columnas, así como el tamaño de nuestros subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 5))

# Eje 0 - Gráfico 1
ax = axs[0]

# Agrupamos por el código de cada película y obtenemos el número de reseñas que recibió cada una
# Obtenemos el recuento de valores únicos del número de reseñas, cuántas películas recibieron un determinado número de reseñas
# Trazamos un gráfico de barras
dft = df_reviews.groupby('tconst')['review'].count() \
    .value_counts() \
    .sort_index()
dft.plot.bar(ax=ax)
ax.set_title('Gráfico de barras de #Reseñas por película')

# Eje 1 - Gráfico 2
ax = axs[1]

# Agrupamos por el identificador de cada película y obtenemos el número de reseñas
# Trazamos un gráfico de distribución de densidades
dft = df_reviews.groupby('tconst')['review'].count()
sns.kdeplot(dft, ax=ax)
ax.set_title('Gráfico KDE de #Reseñas por película')

fig.tight_layout()
```

Se evidencia un mayor número de películas que presentan pocas reseñas y pocas películas con varias reseñas. Es así que más de 1200 películas recibieron una sola reseña y cerca de 1000 recibieron tan solo dos reseñas. Este comportamiento en la distribución de los datos se mantiene a medida que incrementan el número de reseñas, sin embargo existen cerca de 400 películas que llegaron a registrar 30 reseñas. El mismo patrón se registra al analizar el gráfico de distribución de densidades, la densidad de frecuencia más alta de cerca de 0.12 se observa entre las 0 a 5 reseñas, y luego se presenta un nuevo pico alrededor de las 30 reseñas por película.

## Procedimiento de evaluación

Para evaluar el rendimiento de los modelos de aprendizaje automático para texto utilizaremos el valor F1 como métrica de medida, seleccionando aquel modelo que alcance un F1 de al menos 0.85. Construiremos la función `evaluate_model` que permitirá componer una rutina de evaluación que se pueda usar para todos los modelos en este proyecto.

```python
def evaluate_model(model, train_features, train_target, test_features, test_target):

    '''
    Devuelve los valores de exactitud, precisión promedio, puntaje F1 y ROC-AUC.
    También devuelve gráficos de curva ROC-AUC, F1 y PRC (precision-recall curve)

    :model: modelo previamente entrenado y con ajuste de hiperparámetros
    :train_features: características de entrenamiento
    :train_target_ características de prueba
    :test_features: objetivo de entrenamiento
    :test_target: objetivo de prueba

    '''

    eval_stats = {}

    fig, axs = plt.subplots(1, 3, figsize=(20, 6))

    for type, features, target in (('train', train_features, train_target), ('test', test_features, test_target)):

        eval_stats[type] = {}

        pred_target = model.predict(features)
        pred_proba = model.predict_proba(features)[:, 1]

        # F1
        f1_thresholds = np.arange(0, 1.01, 0.05)
        f1_scores = [metrics.f1_score(target, pred_proba>=threshold) for threshold in f1_thresholds]

        # ROC
        fpr, tpr, roc_thresholds = metrics.roc_curve(target, pred_proba)
        roc_auc = metrics.roc_auc_score(target, pred_proba)
        eval_stats[type]['ROC AUC'] = roc_auc

        # PRC
        precision, recall, pr_thresholds = metrics.precision_recall_curve(target, pred_proba)
        aps = metrics.average_precision_score(target, pred_proba)
        eval_stats[type]['APS'] = aps

        if type == 'train':
            color = 'blue'
        else:
            color = 'green'

        # Valor F1
        ax = axs[0]
        max_f1_score_idx = np.argmax(f1_scores)
        ax.plot(f1_thresholds, f1_scores, color=color, label=f'{type}, max={f1_scores[max_f1_score_idx]:.2f} @ {f1_thresholds[max_f1_score_idx]:.2f}')
        # establecer cruces para algunos umbrales
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(f1_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'
            ax.plot(f1_thresholds[closest_value_idx], f1_scores[closest_value_idx], color=marker_color, marker='X', markersize=7)
        ax.set_xlim([-0.02, 1.02])
        ax.set_ylim([-0.02, 1.02])
        ax.set_xlabel('threshold')
        ax.set_ylabel('F1')
        ax.legend(loc='lower center')
        ax.set_title(f'Valor F1')

        # ROC
        ax = axs[1]
        ax.plot(fpr, tpr, color=color, label=f'{type}, ROC AUC={roc_auc:.2f}')
        # establecer cruces para algunos umbrales
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(roc_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'
            ax.plot(fpr[closest_value_idx], tpr[closest_value_idx], color=marker_color, marker='X', markersize=7)
        ax.plot([0, 1], [0, 1], color='grey', linestyle='--')
        ax.set_xlim([-0.02, 1.02])
        ax.set_ylim([-0.02, 1.02])
        ax.set_xlabel('FPR')
        ax.set_ylabel('TPR')
        ax.legend(loc='lower center')
        ax.set_title(f'Curva ROC')

        # PRC
        ax = axs[2]
        ax.plot(recall, precision, color=color, label=f'{type}, AP={aps:.2f}')
        # establecer cruces para algunos umbrales
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(pr_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'
            ax.plot(recall[closest_value_idx], precision[closest_value_idx], color=marker_color, marker='X', markersize=7)
        ax.set_xlim([-0.02, 1.02])
        ax.set_ylim([-0.02, 1.02])
        ax.set_xlabel('recall')
        ax.set_ylabel('precision')
        ax.legend(loc='lower center')
        ax.set_title(f'PRC')

        eval_stats[type]['Exactitud'] = metrics.accuracy_score(target, pred_target)
        eval_stats[type]['F1'] = metrics.f1_score(target, pred_target)

    df_eval_stats = pd.DataFrame(eval_stats)
    df_eval_stats = df_eval_stats.round(2)
    df_eval_stats = df_eval_stats.reindex(index=('Exactitud', 'F1', 'APS', 'ROC AUC'))

    print(df_eval_stats)

    return
```

## Normalización

Para la construcción de modelos predictivos, debemos suponer que todos los modelos que serán entrenados solo aceptan textos en minúsculas, sin dígitos, signos de puntuación o espacios adicionales. Así que vamos a construir la función `clear_text` que se aplicará a la columna `review` y devolverá el texto de cada crítica normalizado.

```python
def clear_text(text):

    '''
    Devuelve un texto normalizado en minúsculas que incluye solo letras y apóstrofos sin espacios
    adicionales.

    :text: texto que será normalizado.

    '''

    pattern = r"[^a-zA-z']" # Establecemos el patrón de texto de solo letras, apóstrofos y espacios
    text = re.sub(pattern, " ", text) # Cambiamos caracteres que no coincidan con el patrón por espacios
    text_1 = text.split() # Obtenemos una lista de palabras
    text_2 = " ".join(text_1) # Recombinamos en una cadena sin espacios adicionales
    return text_2.lower() # Devolvemos texto en minúsculas

# Aplicamos la función 'clear_text' a nuestra columna 'review'
df_reviews['review_norm'] = df_reviews['review'].apply(clear_text)
```

### Modelo 1 - NLTK, TF-IDF y LR

A continuación, entrenaremos nuestro primer modelo, el cual utilizará la librería NLTK (Natural Language Toolkit) para procesar cadenas de texto. Se realizará la tokenización y lematización de cada crítica a través de la creación de la función `nltk_lemmatizer` y luego se aplicará la función a las críticas normalizadas a través de apply.

**PREPROCESAMIENTO DE TEXTO CON NLTK**

```python
def nltk_lemmatizer(text):

    '''
    Devuelve el texto tokenizado y lematizado a partir de la librería NLTK.

    :text: texto que será lematizado

    '''

    lemmatizer = WordNetLemmatizer() # Creamos un objeto de lematización
    tokens = word_tokenize(text) # Dividimos el texto en tokens
    lemmas = [lemmatizer.lemmatize(token) for token in tokens] # Obtenemos los lemas de cada token
    string_lemmas = " ".join(lemmas) # Convertimos la lista de lemas procesados en una línea de texto

    return string_lemmas
```

### Modelo 3 - spaCy, TF-IDF y LR

Ahora cambiaremos el método de preprocesamiento de texto y utilizaremos la librería spaCy para tokenizar y lematizar nuestras críticas de usuarios. Spacy a diferencia de NLTK, tiene un modelo principal para el reconocimiento de texto y utiliza una pipeline especial para procesar el texto entrante. Una vez preparado el texto, vamos a entrenar un modelo de regresión logística.

**PREPROCESAMIENTO DE TEXTO CON SPACY**

```python
# Cargar el modelo de tamaño pequeño en inglés y deshabilitar algunos componentes de la pipeline
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
```

```python
# Definimos una función que ayude al preprocesamiento de texto con spacy
def text_preprocessing_3(text):

    '''
    Devuelve el texto tokenizado y lematizado a partri de la librería spaCy

    :text: texto que sera procesado

    '''
    doc = nlp(text) # Dividimos el texto en tokens
    #tokens = [token.lemma_ for token in doc if not token.is_stop]
    tokens = [token.lemma_ for token in doc] # Obtenemos los lemas de cada token

    return ' '.join(tokens) # Convertimos la lista de lemas procesados en una línea de texto
```

###  Modelo 9 - BERT LR

Hemos trabajado con NLTK y Spacy, ahora vamos a probar con otro método de preprocesamiento, BERT. Este consiste en un modelo de red neuronal para representación de lenguaje, llegando a comprender el contexto de un texto completo y no solo frases cortas. Para el preprocesamiento de texto se utiliza un modelo previamente entrenado en grandes corpus de texto, el cual se encarga de averiguar aquellas palabras desconocidas o enmascaradas y si existe una relación entre las palabras del texto.

A continuación, realizaremos el preprocesamiento de texto con BERT y posteriormente entrenaremos un modelo de regresión logísticas y otro de potenciación de gradiente.

```python
# Iniciamos el tokenizador con una instancia BertTokenizer() con el nombre del modelo previamente entrenado
tokenizer = transformers.BertTokenizer.from_pretrained('bert-base-uncased')

# Iniciamos la configuración BertConfig() con la descripción de la configuración del modelo
config = transformers.BertConfig.from_pretrained('bert-base-uncased')

# Inicamos el modelo BertModel() y pasamos el archivo con el modelo previamente entrenado y su configuración
model = transformers.BertModel.from_pretrained('bert-base-uncased')
```

Construiremos la función `BERT_text_to_embeddings` que devolverá insertados de palabras BERT de 512 palabras y con un tamaño de lotes de 100. Ya que BERT requiere gran potencia computacional y demoraría varias horas obtener insertados del conjunto completo, los datos serán preprocesados en Google Colab y luego se cargará el conjunto de entrenamiento y prueba obtenido.

```python
def BERT_text_to_embeddings(texts, max_length=512, batch_size=100, force_device=None, disable_progress_bar=False):

    '''
    Devuelve insertados de palabras BERT

    :texts: texto del cual se obtendrá insertados
    :max_lenght: longitud máxima de cada vector
    :batch_size: tamaño del lote para creación de insertados
    :force_device: dispositivo que se utilizara, CUDA o CPU
    :disable_progress_bar: activar o desactivar barra de progreso

    '''
    # Generamos listas para ids y attetion masks
    ids_list = []
    attention_mask_list = []

    # Tokenizamos el texto y añadimos los ids y máscaras de atención a las listas correspondientes
    for input_text in texts:
        ids = tokenizer.encode(input_text.lower(), add_special_tokens=True, truncation=True, max_length=max_length)
        padded = np.array(ids + [0]*(max_length - len(ids))) # Rellenamos con cero los vectores
        attention_mask = np.where(padded != 0, 1, 0) # Creamos una máscara para tokens importantes
        ids_list.append(padded)
        attention_mask_list.append(attention_mask)

    # Establecemos el dispositivo que utilizaremos
    if force_device is not None:
        device = torch.device(force_device)
    else:
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # Definimos la barra de progreso
    model.to(device)
    if not disable_progress_bar:
        print(f'Uso del dispositivo {device}.')

    # Obtenemos los insertados de lotes a partir de un bucle, en este caso el tamaño de lote será de 100
    embeddings = []

    for i in tqdm(range(math.ceil(len(ids_list)/batch_size)), disable=disable_progress_bar):

        # Transformamos los datos en un formato tensor, vector multidimensional de Torch
        ids_batch = torch.LongTensor(ids_list[batch_size*i:batch_size*(i+1)]).to(device)
        attention_mask_batch = torch.LongTensor(attention_mask_list[batch_size*i:batch_size*(i+1)]).to(device)

        # Pasamos los datos y máscara al modelo para obtener insertados para el lote
        with torch.no_grad():
            model.eval()
            batch_embeddings = model(input_ids=ids_batch, attention_mask=attention_mask_batch)
        embeddings.append(batch_embeddings[0][:,0,:].detach().cpu().numpy()) # Extraemos elementos de tensor

    # Obtenemos matriz de características
    return np.concatenate(embeddings)
```
